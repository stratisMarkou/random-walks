```python
import numpy as np
from scipy.integrate import ode as ODE

from matplotlib import pyplot as plt
```

```python
f = lambda t, y : np.exp(-50 * y)

ode = ODE(f=f)

y = []

for i in range(100):
    
    ode.set_initial_value(y=0, t=0)
    y.append(ode.integrate(t=(i / 100)))
    
plt.plot(y)
```
